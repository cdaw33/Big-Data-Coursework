```python
#Import the pandas library
import pandas as pd
```

```python
#Import all the databases that will be used throughout the program
test1=pd.read_csv("E:\Big Data Landscape\Datasets\crime.csv")
test2=pd.read_csv("E:\Big Data Landscape\Datasets\education.csv")
test3=pd.read_csv("E:\Big Data Landscape\Datasets\housing.csv")
test4=pd.read_csv("E:\Big Data Landscape\Datasets\house-price-2012.csv", nrows=693)
```

```python
#Joins the Crime and Education databases together using the datazone field
test1=test1[["Datazone","crime_deprivation_rank_2012"]]
test1.columns=["Datazone","Crime"]
test2=test2[["Datazone","education_deprivation_rank_2012"]]
test2.columns=["Datazone","Education"]
dataJoin = pd.merge(test1, test2, on='Datazone')
```

```python
#Joins the Housing database on to the previously created database
dataJoin=dataJoin[["Datazone","Crime","Education"]]
dataJoin.columns=["Datazone","Crime","Education"]
test3=test3[["Datazone","housing_deprivation_rank_2012", "Intermediate geography name"]]
test3.columns=["Datazone","Housing", "Area"]
```

```python
dataFinal = pd.merge(dataJoin, test3, on='Datazone')
#Sorts the database by crime statistics
dataFinal.sort_values(by='Crime', ascending=True)
```

```python
#Imports the fuzz library from the fuzzywuzzy set
from fuzzywuzzy import fuzz
```

```python
#Creates function used to match areas
def match_area(name, list_names, min_score=0):
    # start with -1 encase there is no matches
    max_score = -1
    # Returning empty name for no match as well
    max_name = ""
    # For Loop to move over all the names of the areas
    for name2 in list_names:
        #Finding fuzzy match score
        score = fuzz.ratio(name, name2)
        # Checking if we are above our threshold and have a better score
        if (score > min_score) & (score > max_score):
            max_name = name2
            max_score = score
    return (max_name, max_score)
```

```python
# List for dicts for easy dataframe creation
dict_list = []
# iterating over all the names of areas in the house price database
for name in test4.Area:
    # Use our method to find best match, we can set a threshold here
    match = match_area(name, dataFinal.Area, 50)
    
    # New dict for storing data
    dict_ = {}
    dict_.update({"Area" : name})
    dict_.update({"match_Area" : match[0]})
    dict_.update({"score" : match[1]})
    dict_list.append(dict_)
    
merge_table = pd.DataFrame(dict_list)
# Display results
merge_table
```

```python
#Joins the created database witht the house price database using the previous record matching
merge_table = merge_table[["Area","match_Area"]]
merge_table.columns = ["Area","Match_Area"]
test4=test4[["Area","Price"]]
test4.columns=["Area","Price"]
mergeJoin = pd.merge(merge_table,test4, on='Area' )
dataFinal=dataFinal[["Crime","Education","Housing","Area"]]
dataFinal.columns=["Crime","Education","Housing","Match_Area"]
finalJoin = pd.merge(mergeJoin,dataFinal, on='Match_Area')
finalJoin
```
